```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("e.csv", low_memory=False)
```

# Codes #
## S2AQ5B - How often drank beer last 12 months ##
- 1: Every day
- 2: Nearly every day
- 3: 3 to 4 times a week
- 4: 2 times a week
- 5: Once a week
- 6: 2 to 3 times a month
- 7: Once a month
- 8: 7 to 11 times in the last year
- 9: 3 to 6 times in the last year
- 10: 1 or 2 times in the last year
- 99: Unknown
- BL: NA, did not drink or unknown if drank beer in last 12 months

## S2BQ2E - Number of episodes of alcohol dependence ##

- 1-98: Number of episodes
- 99: Unknown
- BL: NA, lifetime abstainer; did not meet symptom and/or duration criteria for lifetime alcohol dependence

```python
print "The blank space refers to \"BL\""
print "Number of NAs in AGE ",
print len(data["AGE"][data["AGE"].isnull()])
print "Number of NAs in S2BQ2E ",
print len(data["S2BQ2E"][data["S2BQ2E"].isnull()])
print "Number of NAs in S2AQ5B ",
print len(data["S2AQ5B"][data["S2AQ5B"].isnull()])

print "\nCount for S2AQ5B - How often drank beer last 12 months"
print data["S2AQ5B"].value_counts(sort=False)

print "\nPercentages for S2AQ5B - How often drank beer last 12 months"
print data["S2AQ5B"].value_counts(sort=False, normalize=True)

print "\n\nCount for S2BQ2E - Number of episodes of alcohol dependence"
print data["S2BQ2E"].value_counts(sort=False, normalize=False)
print "\nPercentages for S2BQ2E - Number of episodes of alcohol dependence"
print data["S2BQ2E"].value_counts(sort=False, normalize=True)

print "\n\n"
data["S2BQ2E"].convert_objects(convert_numeric=True).hist(bins=len(data["S2BQ2E"].value_counts(sort=False, normalize=False).values))
plt.ylabel("Count")
plt.xlabel("Response")
print "We can se the majority class is for BL"
plt.show()

print "\n\nCount for AGE"
print data["AGE"].value_counts(sort=False, normalize=False)
print "\nPercentages for AGE"
print data["AGE"].value_counts(sort=False, normalize=True)

print "\n"
print "Selecting a subset of rows. "
print "Percentages for S2AQ5B - How often drank beer last 12 months"
print data["S2AQ5B"][data["AGE"] > 50].value_counts(sort=False, normalize=True)
```
